# Importing Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

# Giving image path

```python
img_path = "/content/Student.jpg"
```

# Reading the Image using Numpy array

```python
img = cv2.imread(img_path)
```

```python
# For resizing the Image
img = cv2.resize(img, (100, 100))
```

```python
img
```

# Converting to RGB and GrayScale Format from BGR Format

```python
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
# For resizing the Image
resized_img_rgb = cv2.resize(img_rgb, (100, 100))
```

```python
resized_img_rgb
```

```python
# For resizing the Image
resized_img_gray = cv2.resize(img_gray, (100, 100))
```

```python
resized_img_gray
```

# Displaying the Images

```python
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.imshow(img_rgb)
plt.title("Original Image (RGB)")
plt.axis("off")

plt.subplot(1, 2, 2)
plt.imshow(img_gray, cmap='gray')
plt.title("Grayscale Image")
plt.axis("off")

plt.tight_layout()
plt.show()
```
