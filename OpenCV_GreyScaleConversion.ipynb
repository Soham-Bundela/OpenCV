```python
# import opencv
import cv2
from google.colab.patches import cv2_imshow

# Load the input image
image = cv2.imread('/content/SohamBundela.jpg')

cv2_imshow(image)

# Use the cvtColor() function to grayscale the image
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

cv2_imshow(gray_image)
```

# Method-3

```python
# Import opencv
import cv2
from google.colab.patches import cv2_imshow

# Load the input image
img = cv2.imread('/content/SohamBundela.jpg')

# Check if the image was loaded successfully
if img is not None:
    # Convert the image to grayscale using cv2.cvtColor
    gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

    cv2_imshow(gray_img)
    # cv2.waitKey(0) # waitKey is not needed with cv2_imshow in Colab
    # cv2.destroyAllWindows() # destroyAllWindows is not needed with cv2_imshow in Colab
else:
    print("Error: Image not loaded. Please check the file path.")
```
